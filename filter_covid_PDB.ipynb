```python
import pandas as pd
import numpy as np
import math
```

```python
covid_df = pd.read_csv('rcsb_pdb_custom_report_20211103172321.csv')
```

```python
covid_df
```

```python
# if Entry ID or PDB ID is NaN, will be deleted.
covid_df.drop(covid_df[covid_df['Entry ID'].isnull() | covid_df['PDB ID'].isnull()].index,
    axis=0, inplace=True)
```

```python
# filter out other organism
covid_df = covid_df[covid_df['Source Organism']=='Severe acute respiratory syndrome coronavirus 2']
```

```python
covid_df
```

```python
# remove tuples in column 'Resolution (Å)'. 
temp = []
for i in covid_df['Resolution (Å)'].values:
    
    if not pd.isna(i):
        com_index = i.find(',')

        if com_index > 0:
            first_part = i[:com_index]
            second_part = i[com_index+1:len(i)]

            i = first_part
    
    temp.append(i)
```

```python
covid_df['resolution'] = temp
```

```python
covid_df['resolution'] = covid_df['resolution'].astype(float)
covid_df.info()
```

```python
new_df = covid_df.groupby(by= ['Database Name', 'Accession Code(s)']).agg({'resolution':'min'})
new_df
```

```python
new_df.reset_index(inplace=True)
new_df
```

```python
pd.merge(new_df,covid_df,how='left', on=['Database Name', 'Accession Code(s)', 'resolution'])
```

```python
gene_df = covid_df.groupby(by= 'Gene Name').agg({'resolution':'min'})
gene_df.reset_index(inplace=True)
gene_df
```

```python
pd.merge(gene_df,covid_df,how='left', on=['Gene Name', 'resolution'])
```
